```python
# from station_weather.csv: sort the data and on a single graph, have 3 line plots - temperature and whatever other variable
# filter data across stations.  separate across hours and apply mean function
# each line should look different: different color/style with points marked
# identify max and min on the graph for each line
# have each line in a legend
# for one weather station, there'll be 8 values
# take the mean for every hour (across 3 days) and the mean will be the value
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('station_weather.csv')
fig, ax1 = plt.subplots(figsize=(9, 5))

avg_temp = (df.groupby(["station_id", "hour"])["temperature_f"].mean().reset_index()) # groups rows by station and hour so that the temperatures can be averaged

linestyles = {
    "RISE-01": "-",
    "RISE-02": "--",
    "RISE-03": ":"
}

markers = {
    "RISE-01": "o",
    "RISE-02": "s",
    "RISE-03": "^"
}


for station in avg_temp["station_id"].unique():
    station_data = avg_temp[avg_temp["station_id"] == station]

    plt.plot(station_data["hour"],station_data["temperature_f"],linestyle=linestyles[station],
        marker=markers[station],label=station)



# legend box
ax1.legend(loc='upper left')

plt.title("Average Temperature by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Average Temperature")
plt.xticks(np.arange(6, 21, 2)) # going by hour
plt.grid(True)
plt.legend(title="Station")
plt.tight_layout()
plt.show()
```

```python
# bar charts by each station, by each day: day 1, 2, 3 by rise01, rise02, rise03 - combining all the hours to get one value per day
```

```python
df = pd.read_csv('station_weather.csv')
fig, ax1 = plt.subplots(figsize=(9, 5))

avg_daily = (df.groupby(["station_id", "date"])["temperature_f"].mean().reset_index())

stations = avg_daily["station_id"].unique()
dates = avg_daily["date"].unique()

x=np.arange(len(stations))
width = 0.25

plt.figure(figsize=(8,5))

for i, date in enumerate(dates):
  temps = []

  for station in stations:
    avg = avg_daily[
        (avg_daily["station_id"] == station) &
        (avg_daily["date"] == date)]["temperature_f"].values[0]

    temps.append(avg)

  plt.bar(x + i*width, temps, width=width, label=date)

plt.xticks(x + width, stations)
plt.xlabel("Station")
plt.ylabel("Avg temperature")
plt.title("Avg daily temperature by station")
plt.legend(title="Date")
plt.tight_layout()
plt.show()
```
