```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gme.csv")
df = df.drop(df.index[0])  # 

df["Close"] = df["Close"].astype(float)

plt.figure(figsize=(10, 4))
plt.plot(df.index, df['Close'], marker='.')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Closing Price Over Time')
plt.tight_layout()
plt.show()
```

```python
weekdayavg = np.array

filteredf = df[df['Weekday'] == 'Monday']
monday_avg = filteredf['Return'].mean()
weekday_avg = np.array(monday_avg)
print(monday_avg)

filteredf = df[df['Weekday'] == 'Tuesday']
tuesday_avg = filteredf['Return'].mean()
weekday_avg = np.append(weekday_avg, tuesday_avg)
print(tuesday_avg)

filteredf = df[df['Weekday'] == 'Wednesday']
wednesday_avg = filteredf['Return'].mean()
weekday_avg = np.append(weekday_avg, wednesday_avg)
print(wednesday_avg)

filteredf = df[df['Weekday'] == 'Thursday']
thursday_avg = filteredf['Return'].mean()
weekday_avg = np.append(weekday_avg, thursday_avg)
print(thursday_avg)

filteredf = df[df['Weekday'] == 'Friday']
friday_avg = filteredf['Return'].mean()
weekday_avg = np.append(weekday_avg, tuesday_avg)
print(friday_avg)

max = np.argmax(weekday_avg)

print(weekday_avg)

print("Max: " , max)

min = np.argmin(weekday_avg)
print("Min: " , min)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gme.csv")
df = df.drop(df.index[0])
df["Close"] = pd.to_numeric(df["Close"], errors="coerce")
df = df.dropna(subset=["Close"])

cl = ((df["Close"] * 100) % 10).astype(float).to_numpy()
print(cl)

arr = []
for i in range(10):
    temp = cl[cl == i]
    arr.append(temp.size / cl.size)

print(arr)
```
