## 3. Variables & Data Types
Python is **dynamically typed** ‚Äì variable types are inferred at runtime.

| Type | Example | Literal |
|------|---------|---------|
| `int` | whole numbers | `x = 42` |
| `float` | decimal numbers | `pi = 3.14159` |
| `str` | text | `name = "Ada"` |
| `bool` | truth values | `is_valid = True` |
| `list` | ordered, mutable collection | `scores = [95, 88, 72]` |
| `dict` | key‚Äëvalue pairs | `user = {'id': 1, 'name': 'Sam'}` |

Run the cell below to see the types in action.

```python
x = 42
pi = 3.14159
name = "Ada"
is_valid = True
scores = [95, 88, 72]
user = {'id': 1, 'name': 'Sam'}

print(type(x), type(pi), type(name), type(is_valid), type(scores), type(user))
```

## Activity 4 ‚Äì Factorial Function
Implement `factorial(n)` that returns the factorial of n (`factorial(0)=1`, `factorial(1)=1`, `factorial(2)=2`, etc.). Test that `factorial(5)` is 120.

   Implement `is_prime(n)` efficiently for `n ‚â§ 10^6`. Test it on the following primes: 2, 3, 5, 7, 9973

Choose any you like and work below.

```python
#1
import timeit

forloop_time = timeit.timeit(
    stmt="""
squares = []
for x in range(1, 100001):
    squares.append(x**2)
""",
    number=1
)

print(forloop_time)

listcomp_time = timeit.timeit(stmt = "[x**2 for x in range(1, 100001)]", number = 1)
print(listcomp_time)
```

```python
#4
import math
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False

    return True

print(is_prime(2))
print(is_prime(3))
print(is_prime(5))
print(is_prime(7))
print(is_prime(9973))
```
